```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import statistics as st
```

```python
n = 1000 # capacity
iterations = int(n)
```

```python
data_set_0 = np.random.rand(iterations) # sample
print('Matrix: ', data_set_0)
```

```python
median = np.median(data_set_0)
print('Expected value: ', median)
```

```python
mean = st.mean(data_set_0)
print('Mean: ', mean)
```

```python
variance = np.var(data_set_0)
print('Variance: ', variance)
```

```python
standart_deviation = mt.sqrt(variance)
print('Standart deviation: ', standart_deviation)
```

```python
bins = 20
plt.hist(data_set_0, bins=bins, color='c', edgecolor='k', alpha=0.5)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
n = 10000  # capacity
iterations = int(n)

data_set_1 = np.random.rand(iterations)
print('Matrix: ', data_set_1)

median = np.median(data_set_1)
print('Expected value: ', median)

mean = st.mean(data_set_1)
print('Mean: ', mean)

variance = np.var(data_set_1)
print('Variance: ', variance)

standart_deviation = mt.sqrt(variance)
print('Standart deviation: ', standart_deviation)

bins = 20
plt.hist(data_set_1, bins=bins, color='c', edgecolor='k', alpha=0.5)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
rand_arr =np.zeros(n)
for i in range(n):
    rand_arr[i]=np.random.randint(-50, 50)
```

```python
model = np.zeros(n)
measured_process = np.zeros(n)
for i in range(n):
    model[i]= 1 # constant model of a real process
    measured_process[i] = model[i] + rand_arr[i]

plt.xlim(0, 100)
plt.ylim(-80, 80)
plt.plot(measured_process, color="blue", linestyle="solid", marker='o')
plt.plot(model, color='green')
```

```python
dm = 0
dsig = 5
```

```python
data_set_2 = np.random.normal(dm, dsig, n) # sample # normal distribution with parameters
print('Matrix: ', data_set_2)

data_set_3 = np.random.randn(n) # random normal distribution w/o parameters
```

```python
median = np.median(data_set_2)
print('Expected value: ', median)
```

```python
mean = st.mean(data_set_2)
print('Mean: ', mean)
```

```python
variance = np.var(data_set_2)
print('Variance: ', variance)
```

```python
standart_deviation = mt.sqrt(variance)
print('Standart deviation: ', standart_deviation)
```

```python
plt.hist(data_set_2, bins=20, facecolor="blue", edgecolor='k', alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
model = np.zeros(n)
measured_process = np.zeros(n)

for i in range(n):
    model[i]=(0.000001*i*i)  # quadratic model of a real process
    measured_process[i] = model[i] + data_set_2[i]
```

```python
plt.plot(measured_process, color='blue', label='Real') #  
plt.plot(model, color='green', label='Model') #  
plt.legend()
plt.show()
```

```python
plt.hist(data_set_2, bins=250, alpha=0.5, color='blue', label='Random')
plt.hist(data_set_3, bins=250, alpha=0.5, color='red', label='Standart Normal Distribution')
plt.hist(measured_process, bins=250, alpha=0.5, color='green', label='Measured')
plt.legend()
plt.show()
```
